```python
import tensorflow as tf
import numpy as np
from tensorflow import keras
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import random
```

```python
# инициализирую данные
x = 0.5
siz = 100000
```

```python
data = np.zeros((siz, 5))
```

```python
# генерирую случайным образом параметры для расчета данных для обучения
for i in range(0, siz):
  n = random.randint(-10, 10)
  m = random.randint(-10, 10)
  l = random.randint(-10, 10)
  k = random.randint(-10, 10)
  j = random.randint(-10, 10)
  data[i, 0] = n
  data[i, 1] = m
  data[i, 2] = l
  data[i, 3] = k
  data[i, 4] = j
```

```python
# удаляю повторяющеяся элементы
unique_data = np.unique(data, axis = 0)
shape = unique_data.shape
print(shape[0])

```

```python
oute = np.zeros((shape[0], 1))
```

```python
# расчет данных выборки
from re import X
for i in range(0, shape[0]):
  oute[i, 0] = (x ** (unique_data[i, 0])) + (unique_data[i, 1] * np.sin(x)) - (np.sqrt((x ** (-unique_data[i, 2])))) + (unique_data[i, 3] * x) - (unique_data[i, 4] / x)
# print(unique_data)
```

```python
# Разделяем данные на обучающую и тестовую выборки
X_train, X_test, y_train, y_test = train_test_split(unique_data, oute, test_size = 0.2, random_state = 22)
```

```python
# реализация нейронной сети
model = tf.keras.models.Sequential([
  tf.keras.layers.Dense(25, activation='tanh', input_shape=(5,)),
  tf.keras.layers.Dense(10, activation='tanh'),
  tf.keras.layers.Dense(10, activation='tanh'),
  tf.keras.layers.Dense(10, activation='tanh'),
  tf.keras.layers.Dense(10, activation='tanh'),
  tf.keras.layers.Dense(1)
])

```

```python
model.compile(loss='mse', optimizer = tf.keras.optimizers.Adam())
```

```python
# реализация обучения нейронной сети
model.fit(X_train, y_train, epochs = 1000, batch_size = 100, validation_data=(X_test, y_test))
```

```python
# ошибка модели
loss = model.evaluate(X_test, y_test, batch_size = 100)
print('Test loss:', loss)
```
